```python
import matplotlib.pyplot as plt
import re

df = pd.read_csv()

# Assume the df has the following columns(don't have to be the same name): ['Models', 'label']

Mflag=0         # indicate whether 'Model' exists
Lflag=0         # indicate whether 'label' exists

for col in df.columns:
    if re.search(r'[Mm]odels?', col): # just for example, can be changed to dataset or other
        Mflag=1
        Models = df[col]

    elif re.search(r'[Ll]abels?', col):
        Lflag=1
        Labels = df[col]

if not (Mflag and Lflag):
    raise KeyError("Missing columns.")

combined = pd.concat([Models, Labels], axis=1)
grouped = combined.groupby(combined.columns[0]) # which is different models

model, entail, neural, contra, unk = [],[],[],[],[]

for m, sub_df in grouped: # for every model, align the different label to it
    model.append(m)
    entail.append(sub_df.loc[sub_df[sub_df.columns[1]]==0].shape[0])
    neural.append(sub_df.loc[sub_df[sub_df.columns[1]]==1].shape[0])
    contra.append(sub_df.loc[sub_df[sub_df.columns[1]]==2].shape[0])
    unk.append(sub_df.loc[sub_df[sub_df.columns[1]]==3].shape[0])



################################################################
# Above is data processing
# Below is plotting
################################################################


fig, ax = plt.subplots()  # remain a way to draw nulti-plots

# stacked bar chart
# color can use RGB or name
ax.bar(model, entail, label='Entailment', color='green', width=0.5)
ax.bar(model, neural, bottom=entail, label='Neural', color='orange', width=0.5)
ax.bar(model, contra, bottom=[i+j for i,j in zip(entail, neural)], label='Contradiction', color='blue', width=0.5)
ax.bar(model, unk, bottom=[i+j+k for i,j,k in zip(entail, neural, contra)], label='UNK', color='purple', width=0.5)

# Grouped histogram
x = np.arange(len(model))
width = 0.3 

ax.bar(x - 1.5*width, entail, width, label='Entailment')
ax.bar(x - 0.5*width, neural, width, label='Neural')
ax.bar(x + 0.5*width, contra, width, label='Contradiction')
ax.bar(x + 1.5*width, unk, width, label='UHK')
ax.set_xticklabels(model)

# plot
ax.plot(x=, y=, label='Entailment', color='green', width=0.5, linestyle='-',linewidth=0.2, marker='o', markercolor='blue')


# label and legend
ax.set_xlabel('Models', fontsize=14, fontdict={'family': 'serif', 'color':  'black'})
ax.set_ylabel('Count of each category', fontsize=14, fontdict={'family': 'serif', 'color':  'black'})

ax.set_title('bar chart', fontsize=14, fontdict={'family': 'serif', 'color':  'black'})

ax.tick_params(axis='both', labelsize=12,labelcoloe='green',color='blue')

# discrete way
# ax.set_xticks([])
# ax.set_yticks([])

# continuous way
ax.set_xlim(1, 3) # min to max
ax.set_ylim(0, 10)

ax.set_title('My Plot', fontsize=16, fontdict={'family': 'serif','color':'red'})

ax.legend(loc=(0.2, 0.7)) # or use like 'center', 'upper lest' or just leave it blank for auto-adjustment

plt.show()
```
